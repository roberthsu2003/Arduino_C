## 數學運算


```python
# 建立變數

10 #整數(int)
10.0 #浮點數(float)
"10.5" # 字串(str)
'10.5' # 字串(str)
```

```python
print(10 + 5) # 敘述式(statement)

print(10.0 + 5.0)

print("10.0 + 5.0")

print(10.0 + 5)
```

```python
10 + 5 #運算式(expression)
```

```python
a = 5
b = 10
print(a)
print(b)
print(a + b)
```

```python
a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.3333333333333335
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 1000
```
